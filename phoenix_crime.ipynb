Phoenixopendata.com uses CKAN API. I found a python wrapper for it named ckanapi. So we're gonna give that a try...

```python
import ckanapi
from ckanapi import RemoteCKAN
```

Had to register for a developer account to get an API Key.
https://www.phoenixopendata.com/user/register

```python
my_key = ''
```

Now to connect to the API and see what they've got.

```python
ckan = RemoteCKAN('https://www.phoenixopendata.com', apikey=my_key)
ckan.action.package_list()
```

```python
ckan.action.package_show(id='crime-data')
```

```python
ckan.action.package_activity_list(id='crime-data')
```

```python
ckan.action.datastore_info('crime-data')
```

```python
ckan.action.datastore_resource_id_from_alias('crime-data')
```

```python
ckan.action.resource_show('crime-data')
```

```python
ckan.call_action('datastore_info', {'resrouce_id': 'crime-data'})
```

```python
ckan.call_action('resource_show', {'id': 'crime-data'})
```

```python
ckan.call_action('datastore_info', {'id': '0ce3411a-2fc6-4302-a33f-167f68608a20'})
```
